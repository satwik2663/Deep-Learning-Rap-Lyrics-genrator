## Jaccard Similarity

```python
def get_jaccard_sim(text1, text2): 
    a = set(text1.split()) 
    b = set(text2.split())
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))
```

```python
print('Jaccard similarity is: ', get_jaccard_sim(data,data2))
```
